### Criando uma função e aplicando em uma coluna inteira para gerar uma nova coluna

```python
def duplicaBolsa(rows):
    return rows*2
```

```python
df['nova_bolsa'] = df.bolsa_integral_ampla.apply(duplicaBolsa)
df.head()
```

### Aplicando uma função lambda em uma coluna para gerar uma nova coluna

```python
df['nova_bolsa2'] = df.bolsa_integral_ampla.apply(lambda x : x * 100)
df.head()
```
